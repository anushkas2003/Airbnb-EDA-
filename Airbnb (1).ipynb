```python
import streamlit as st
import pandas as pd
import plotly.express as px

# Load Data
data_path = 'Airbnb.csv'  # Ensure this file is in the same directory
df = pd.read_csv(data_path, encoding_errors='ignore')

# Streamlit UI
st.set_page_config(page_title="Airbnb NYC Dashboard", layout="wide")
st.title("🏡 Airbnb Listings Dashboard - New York")

# Sidebar Filters
st.sidebar.header("Filters")
nb_filter = st.sidebar.multiselect("Select Neighborhood", options=df["neighbourhood_group"].unique(), default=df["neighbourhood_group"].unique())
price_range = st.sidebar.slider("Price Range", min_value=int(df["price"].min()), max_value=int(df["price"].max()), value=(50, 500))

# Filter Data
df_filtered = df[(df["neighbourhood_group"].isin(nb_filter)) & (df["price"].between(price_range[0], price_range[1]))]

# Price Distribution
st.subheader("Price Distribution")
fig_price = px.histogram(df_filtered, x="price", nbins=50, color_discrete_sequence=["pink"])
st.plotly_chart(fig_price, use_container_width=True)

# Availability Analysis
st.subheader("Availability in Days")
fig_availability = px.histogram(df_filtered, x="availability_365", nbins=50, color_discrete_sequence=["lightgreen"])
st.plotly_chart(fig_availability, use_container_width=True)

# Map of Listings
st.subheader("Map of Listings")
fig_map = px.scatter_mapbox(df_filtered, lat="latitude", lon="longitude", color="price", size_max=15, zoom=10,
                            mapbox_style="carto-positron", color_continuous_scale="pinkyl")
st.plotly_chart(fig_map, use_container_width=True)

# Summary Stats
st.sidebar.subheader("Summary Statistics")
st.sidebar.write(df_filtered.describe()[["price", "availability_365"]])
```
